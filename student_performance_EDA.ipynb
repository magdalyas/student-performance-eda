## Student Performance Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset Overview

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

## Data Quality Check

```python
print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Records:")
print(df.duplicated().sum())
```

## Average Score Metric

To measure overall student performance, an average score was calculated using math, reading, and writing scores.

```python
df["average score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3
```

```python
df[["math score", "reading score", "writing score", "average score"]].head()
```

# Distribution of Math Scores

```python
sns.histplot(df["math score"], bins=20)
plt.title("Distribution of Math Scores")

plt.xlabel("Math Score")
plt.ylabel("Number of Students")

plt.show()
```

##Observation

Most students scored between 60 and 80 on the math exam, while fewer students received very low or very high scores.

## Impact of Test preparation on Math Scores

```python
df.groupby("test preparation course")["average score"].mean()
```

```python
sns.boxplot(
    x="test preparation course",
    y="math score",
    data=df
)

plt.title("Math Scores by Test Preparation")

plt.show()
```

##Observation

Students who completed the test preparation course generally achieved higher math scores compared to students who did not complete the course.

## Impact of Test Preparation on Reading Scores

```python
sns.boxplot(
    x="test preparation course",
    y="reading score",
    data=df
)

plt.title("Reading Scores by Test Preparation")

plt.show()
```

##Observation

Students who completed the test preparation course also achieved higher reading scores compared to students who did not complete the course.

## Impact of Test Preparation on Writing Scores

```python
sns.boxplot(
    x="test preparation course",
    y="writing score",
    data=df
)

plt.title("Writing Scores by Test Preparation")

plt.show()
```

##Observation

Students who completed the test preparation course also achieved higher writing scores compared to students who did not complete the course.

## Relationship Between Reading and Writing Scores

```python
sns.scatterplot(
    x="reading score",
    y="writing score",
    data=df
)

plt.title("Reading vs Writing Scores")
plt.show()
```

## Observation
Reading and writing scores show a strong positive relationship. Students with higher reading scores generally achieved higher writing scores, suggesting that performance in these subjects is closely connected

# Conclusion

This analysis explored factors related to student academic performance using exam scores and educational variables.

Key findings include:

- Most students scored between 60 and 80 on the math exam.
- Students who completed the test preparation course generally achieved higher scores in math, reading, and writing.
- Reading and writing scores showed a strong positive relationship, suggesting that performance in these subjects is closely connected.
- The dataset contained no missing values or duplicate records, indicating good data quality.

Overall, the results suggest that test preparation may be associated with improved academic performance across multiple subjects.
